```python
import torch
```

```python
a=torch.ones(2,3,3)
print(a)

aa,b,c = a.size()
print(aa)
```

The tensor in the torch is similar as the numpy array. A 3-D tensor means 3 channels or layers 2-D matrix.

```python
x=torch.tensor([6,3], dtype=torch.double)
print(x)
xx=x.new_ones(5,3)
print(xx)
x = torch.randn_like(xx, dtype=torch.float)    # override dtype!
print(x)  
```

可以看到，new_ones函数复用了x的dtype属性，randn_like函数复用了x的shape同时通过手动指定数据类型覆盖了原有的dtype属性

```python
torch.arange(1, 10, 2)
```

arrange函数参数有三个，分别是范围的起始值、范围的结束值和步长

Part2: Operations

```python
'''加减乘除'''
y = torch.rand(5,3)
z1 = xx+y
print(z1)
```

```python
z2 = torch.add(z1,y)
print(z2)
```

```python
z3 = torch.empty(5,3)
print(z3)
torch.add(z1,z2, out=z3)
print(z3)
```

```python
z2.add_(z1)
print(z2)
```

```python
c=torch.sub(z1,z2)
print(c)
d=torch.mul(z1,z2)
print(d)
e=torch.div(z1,z2)
print(e)
```

```python
'''tensor for matrix multiplication'''
a = torch.tensor([[-1,3,4],[2,2,2]])
print(a)
b = a.t()
c = torch.mm(a,b)
print(c)
print(a.abs())
```
